```python
import pandas as pd
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split


data = pd.read_csv('/kaggle/input/star-dataset/6 class csv.csv')

# One-hot encode 
encoder = OneHotEncoder()
star_colors = encoder.fit_transform(data[["Star color"]]).toarray()
star_colors_df = pd.DataFrame(star_colors, columns=encoder.get_feature_names_out(["Star color"]))
data = pd.concat([data, star_colors_df], axis=1).drop(columns=["Star color"], axis=1)
encoder = LabelEncoder()
data["Spectral Class"] = encoder.fit_transform(data["Spectral Class"])

ft = data.drop(columns=["Star type"])
target = data["Star type"]

trainf, testf, traint, testt = train_test_split(ft, target, test_size=0.2, random_state=0)

model = RandomForestClassifier(n_estimators=100, random_state=0)
model.fit(trainf, traint)

pred = model.predict(testf)

acc = accuracy_score(testt, pred)
print("Accuracy: ", acc)

plt.scatter(data["Temperature (K)"], data["Luminosity(L/Lo)"], c=data["Spectral Class"], cmap="viridis")
plt.xlabel("Temperature (K)")
plt.ylabel("Luminosity (L/Lo)")
plt.title("Hertzsprung Russell Diagram")
plt.show()
```

```python
import pandas as pd
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split


data = pd.read_csv('/kaggle/input/star-dataset/6 class csv.csv')

# One-hot encode 
encoder = OneHotEncoder()
star_colors = encoder.fit_transform(data[["Star color"]]).toarray()
star_colors_df = pd.DataFrame(star_colors, columns=encoder.get_feature_names_out(["Star color"]))
data = pd.concat([data, star_colors_df], axis=1).drop(columns=["Star color"], axis=1)
encoder = LabelEncoder()
data["Spectral Class"] = encoder.fit_transform(data["Spectral Class"])

ft = data.drop(columns=["Star type"])
target = data["Star type"]

trainf, testf, traint, testt = train_test_split(ft, target, test_size=0.2, random_state=0)

model = RandomForestClassifier(n_estimators=100, random_state=0)
model.fit(trainf, traint)

pred = model.predict(testf)

acc = accuracy_score(testt, pred)
print("Accuracy: ", acc)

plt.scatter(data["Temperature (K)"], data["Luminosity(L/Lo)"], c=data["Spectral Class"], cmap="viridis")
plt.xlabel("Temperature (K)")
plt.ylabel("Luminosity (L/Lo)")
plt.title("Hertzsprung Russell Diagram")
plt.show()



#PREDICTOR
temperature = float(input("Enter temp (in K): "))
X = [[temperature]]
prediction = model.predict(X)[0]
print("The star is {prediction}")
```
